```python
import pdfplumber
import re
from ics import Calendar, Event
from datetime import datetime
import pytz

# Path to the PDF file
pdf_path = r"C:\Users\ThuNguyen\Desktop\Test_timetable.pdf"

# Function to extract text from the PDF
def extract_text_from_pdf(pdf_path):
    try:
        with pdfplumber.open(pdf_path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text()
        return text
    except FileNotFoundError:
        print("The file was not found.")
        return ""
    except Exception as e:
        print(f"An error occurred: {e}")
        return ""

# Extracted text from the PDF
text = extract_text_from_pdf(pdf_path)

# Print the extracted text (optional)
print("Extracted text from PDF:\n", text)

# Regular expression patterns to extract dates, times, titles, and locations
event_pattern = r"(.+?) (\d{1,2}/\d{1,2}/\d{4}) (\d{1,2}:\d{2}) (\d{1,2}:\d{2}) (.*)"
events = re.findall(event_pattern, text)

# Combine extracted data into a list of meeting dictionaries
meetings = []
for event in events:
    title = event[0].strip()
    date = event[1].strip()
    start_time = event[2].strip()
    end_time = event[3].strip()
    location = event[4].strip()
    
    meeting = {
        "date": date,
        "start_time": start_time,
        "end_time": end_time,
        "title": title,
        "location": location
    }
    meetings.append(meeting)

# Print the list of meetings to ensure data is correct
print("Extracted Meetings Data:\n", meetings)

# Define the timezone (e.g., Europe/Berlin)
timezone_str = 'Europe/Berlin'
timezone = pytz.timezone(timezone_str)

# Create a calendar and add the meetings as events
cal = Calendar()

for meeting in meetings:
    try:
        # Convert date and time to datetime objects with the specified timezone
        start_datetime = datetime.strptime(f"{meeting['date']} {meeting['start_time']}", "%d/%m/%Y %H:%M")
        end_datetime = datetime.strptime(f"{meeting['date']} {meeting['end_time']}", "%d/%m/%Y %H:%M")
        
        # Localize the datetime objects to the specified timezone
        start_datetime = timezone.localize(start_datetime)
        end_datetime = timezone.localize(end_datetime)

        # Creating the event
        event = Event()
        event.name = meeting["title"]
        event.begin = start_datetime
        event.end = end_datetime
        event.location = meeting["location"]
        
        cal.events.add(event)

        # Print event details to ensure they are correct
        print(f"Added Event: {event.name} at {event.location} from {event.begin} to {event.end}")
    except Exception as e:
        print(f"Failed to add event: {meeting['title']} on {meeting['date']} due to error: {e}")

# Check if the calendar has events
if len(cal.events) > 0:
    # Save the calendar as .ics file
    output_path = r"C:\Users\ThuNguyen\Desktop\meetings.ics"
    with open(output_path, "w") as f:
        f.write(str(cal))
    print(f"Calendar has been saved as {output_path}")
else:
    print("No events were added to the calendar.")
```
